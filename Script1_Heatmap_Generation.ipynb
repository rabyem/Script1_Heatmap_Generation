```python
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv("TPM_M1146+ACT Data.csv", index_col='Locus_tag')
df
```

```python
pd.read_csv("SoxR regulon.csv")
```

```python
df1 = df.loc[["SCO0319", "SCO0320", "SCO0321", "SCO1178", "SCO1697", "SCO1734", "SCO1909", "SCO2476", "SCO2477", "SCO2478", "SCO4021", "SCO4264", "SCO4265", "SCO4266", "SCO7008"]]
df1
```

```python
df2 = df1.stack().rename_axis(('Locus tag', 'Time')).reset_index(name='Value')
df2
```

```python
colors = ("black", "white", "red")
lims= (0,50)
b = (5,25,50)
a = ggplot(df2, aes(x='Time', y='Locus tag')) + geom_tile(aes(fill = 'Value'))+ scale_fill_gradientn(limits= lims,colors= colors, breaks=(5,25,50),labels=("5", "25", "50"))
b = a + ggtitle('SOXR REGULON M1146 +ACT') + labs(y="Locus tag", x = "Time")
b
```

```python
b.save('HeatMap1.png', width = 14, height = 14)
```

```python
df3 = df2.copy()
df3.loc[df2["Value"] <= 5, 'Value'] = "x<=5"
df3.loc[df2["Value"] > 5, 'Value'] = "x>5"
df3
```

```python
a = ggplot(df3, aes(x='Time', y='Locus tag')) + geom_tile(aes(fill = 'Value')) + scale_fill_manual(values = ("red", "salmon"))
b = a + ggtitle('SOXR REGULON M1146 +ACT') + labs(y="Locus tag", x = "Time")
b
```

```python
b.save('HeatMap2.png', width = 14, height = 14)
```

```python
df1.transpose()
df4 = df1.transpose()
corr = df4.corr(method='pearson')
corr
```

```python
df5 = corr.stack().rename_axis(('x', 'y')).reset_index(name='Value')
df5
```

```python
colors = ("blue", "cornflowerblue", "white", "pink", "red")
limits=(-1,1)
breaks = (-1, -0.5, 0, 0.5, 1)
a = ggplot(df5, aes(x='x', y='y')) + geom_tile(aes(fill = 'Value'))+ scale_fill_gradientn(limits= limits,colors= colors, breaks=(-1,-0.5,0,0.5,1),labels=("-1", "-0.5", "0", "0.5", "1"))
b = a + ggtitle('SOXR REGULON M1146 +ACT') + labs(y="Locus tag", x = "Locus tag")
c = b + theme(axis_text_x = element_text(angle = 90, hjust=1))
c
```

```python
c.save('HeatMap3.png', width = 14, height = 14)
```
