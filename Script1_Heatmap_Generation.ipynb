```python
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv("TPM_M1146+ACT Data.csv", index_col='Locus_tag')
df
```

```python
df1 = pd.read_csv("SoxR regulon.csv")
df1
```

```python
df2 = df.loc[df1['Locus tag']]
df2
```

```python
df3 = df2.stack().rename_axis(('Locus tag', 'Time')).reset_index(name='Value')
df3
```

```python
colors = ("black", "white", "red")
lims= (0,50)
b = (5,25,50)
a = ggplot(df3, aes(x='Time', y='Locus tag')) + geom_tile(aes(fill = 'Value'))+ scale_fill_gradientn(limits= lims,colors= colors, breaks=(5,25,50),labels=("5", "25", "50"))
b = a + ggtitle('SOXR REGULON M1146 +ACT') + labs(y="Locus tag", x = "Time")
b
```

```python
b.save('HeatMap1.png', width = 14, height = 14)
```

```python
df4 = df3.copy()
df4.loc[df3["Value"] <= 5, 'Value']  = "x<=5"
df4.loc[((df3["Value"] >5) & (df3["Value"] <=25), 'Value')] =  "5<x<=25"
df4.loc[((df3["Value"] >25) & (df3["Value"] <=50), 'Value')] = "25<x<=50"
df4.loc[df3["Value"] > 50, 'Value'] = "x>50"
df4
```

```python
a = ggplot(df4, aes(x='Time', y='Locus tag')) + geom_tile(aes(fill = 'Value')) + scale_fill_manual(values = ("royalblue", "deepskyblue", "slateblue", "gainsboro"))
b = a + ggtitle('SOXR REGULON M1146 +ACT') + labs(y="Locus tag", x = "Time")
b
```

```python
b.save('HeatMap2.png', width = 14, height = 14)
```

```python
df2.transpose()
df5 = df2.transpose()
corr = df5.corr(method='pearson')
corr
```

```python
df6 = corr.stack().rename_axis(('x', 'y')).reset_index(name='Value')
df6
```

```python
colors = ("blue", "cornflowerblue", "white", "pink", "red")
limits=(-1,1)
breaks = (-1, -0.5, 0, 0.5, 1)
a = ggplot(df6, aes(x='x', y='y')) + geom_tile(aes(fill = 'Value'))+ scale_fill_gradientn(limits= limits,colors= colors, breaks=(-1,-0.5,0,0.5,1),labels=("-1", "-0.5", "0", "0.5", "1"))
b = a + ggtitle('SOXR REGULON M1146 +ACT') + labs(y="Locus tag", x = "Locus tag")
c = b + theme(axis_text_x = element_text(angle = 90, hjust=1))
c
```

```python
c.save('HeatMap3.png', width = 14, height = 14)
```
